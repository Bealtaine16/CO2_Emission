2. Filtered unused columns.

```python
df_subset = df[['country', 'year', 'iso_code', 'population', 'gdp', 'primary_energy_consumption', 'cement_co2', 'coal_co2', 'consumption_co2', 
                'flaring_co2', 'gas_co2', 'land_use_change_co2', 'oil_co2', 'other_industry_co2', 'trade_co2', 'co2', 'co2_including_luc']]
df_subset
```

4. Filtered not country rows.

```python
df_subset = df_subset[(~pd.isna(df_subset['iso_code'])) | (df_subset['country'] == 'Kosovo')]
df_subset.loc[df['country'] == 'Kosovo', 'iso_code'] = 'XKK'
df_subset.reset_index(drop=True, inplace=True)
df_subset
```

5. Investigate data description.

```python
df_subset.describe(include='all')
```

6. Change the GDP values to more current ones

```python
df_mpd2023 = pd.read_excel("../datasets/gdp_mpd2023.xlsx", sheet_name="Full data")

df_gdp = df_mpd2023.copy()
df_gdp['gdp'] = df_gdp['gdppc'] * df_gdp['pop'] * 1000
df_gdp = df_gdp.rename(columns={'countrycode': 'iso_code', "country": "country_name"})
df_gdp
```

```python
df_merged = pd.merge(df_subset, df_gdp, on=['iso_code', 'year'])
```

```python
#check differences
df_merged['gdp_diff'] = np.where(
    np.isnan(df_merged['gdp_x']) & np.isnan(df_merged['gdp_y']),
    False,  # Set to False if both are NaN
    df_merged['gdp_x'] != df_merged['gdp_y']
)
df_differences = df_merged[df_merged['gdp_diff']]

df_differences[['country', 'iso_code', 'year', 'gdp_x', 'gdp_y']].to_csv('../datasets/gdp_differences_check.csv', index=False)
```

```python
df_subset = df_merged.drop(columns={'gdp_x', 'country_name', 'region', 'gdppc', 'pop', 'gdp_diff'})
df_subset = df_subset.rename(columns={'gdp_y': 'gdp'})
df_subset
```

7. Save data.

```python
df_subset.to_csv('../output/1_data_exploration.csv', index=False)
```
