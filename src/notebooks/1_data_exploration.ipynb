4. Filtered not country rows.

```python
df_subset = df_subset[(~pd.isna(df_subset['iso_code'])) | (df_subset['country'] == 'Kosovo')]
df_subset.loc[df['country'] == 'Kosovo', 'iso_code'] = 'XKK'
df_subset.reset_index(drop=True, inplace=True)
df_subset
```

6. Check the distribution of CO2 emissions around the world

6.1. Check cumulative co2 emissions (until 2021)

```python
df_co2_cumulative_distribution = df_subset[df_subset['year'] == 2021][['country', 'cumulative_co2_including_luc']]
df_co2_cumulative_distribution
```

```python
#cumulative co2 emissions - %
total_cumulative_emissions = df_co2_cumulative_distribution['cumulative_co2_including_luc'].sum()
df_co2_cumulative_distribution['co2_cumulative_percentage'] = (df_co2_cumulative_distribution['cumulative_co2_including_luc'] / total_cumulative_emissions) * 100
df_co2_cumulative_sorted = df_co2_cumulative_distribution.sort_values(by='co2_cumulative_percentage', ascending=False).reset_index(drop=True)
df_co2_cumulative_sorted
```

```python
# Save in csv
df_co2_cumulative_sorted.to_csv(os.path.join(config.output_exploration, 'co2_distribution_cumulative_all_countries.csv'), index=False)
```

```python
# Identifying the countries responsible for 90% of the emissions
df_co2_cumulative_sorted['co2_cumulative_top_90'] = df_co2_cumulative_sorted['co2_cumulative_percentage'].cumsum()
df_co2_cumulative_top_90 = df_co2_cumulative_sorted[df_co2_cumulative_sorted['co2_cumulative_top_90'] <= 90]
df_co2_cumulative_top_90
```

```python
# Save in csv
df_co2_cumulative_top_90.to_csv(os.path.join(config.output_exploration, 'co2_distribution_cumulative_top_90_countries.csv'), index=False)
```

6.2. Check co2 emissions from 2021

```python
df_co2_2021_distribution = df_subset[df_subset['year'] == 2021][['country', 'co2_including_luc']]
df_co2_2021_distribution
```

```python
#2021 co2 emissions - %
total_2021_emissions = df_co2_2021_distribution['co2_including_luc'].sum()
df_co2_2021_distribution['co2_2021_percentage'] = (df_co2_2021_distribution['co2_including_luc'] / total_2021_emissions) * 100
df_co2_2021_sorted = df_co2_2021_distribution.sort_values(by='co2_2021_percentage', ascending=False).reset_index(drop=True)
df_co2_2021_sorted
```

```python
# Save in csv
df_co2_2021_sorted.to_csv(os.path.join(config.output_exploration, 'co2_distribution_all_countries.csv'), index=False)
```

```python
# Identifying the countries responsible for 90% of the emissions
df_co2_2021_sorted['co2_2021_top_90'] = df_co2_2021_sorted['co2_2021_percentage'].cumsum()
df_co2_2021_top_90 = df_co2_2021_sorted[np.round(df_co2_2021_sorted['co2_2021_top_90']) <= 90]
df_co2_2021_top_90
```

```python
# Save in csv
df_co2_2021_top_90.to_csv(os.path.join(config.output_exploration, 'co2_distribution_top_90_countries.csv'), index=False)
```

7. Check the data for the  45 countries that have the largest cumulative co2 emissions and are responsible for 90% of all total co2 emissions
